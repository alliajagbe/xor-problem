```python
import numpy as np
from tensorflow import keras
from keras.layers import Dense
import matplotlib.pyplot as plt
```

```python

data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
target = np.array([0, 1, 1, 0])
```

```python
model = keras.Sequential()
model.add(Dense(8, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
learning_rate = 0.1
sgd = keras.optimizers.SGD(learning_rate=learning_rate)

model.compile(optimizer=sgd, loss='mean_squared_error')
```

```python
from keras.callbacks import Callback

class LearningRateRecorder(Callback):
    def on_epoch_end(self, epoch, logs=None):
        current_learning_rate = float(self.model.optimizer.learning_rate)
        lr.append(current_learning_rate)

# Initialize the parameters
converged = False
t_loss = 0.001
loss_fns = []
lr = []

# Training loop
while not converged:
    history = model.fit(data, target, epochs=10, verbose=0, callbacks=[LearningRateRecorder()])
    loss = history.history['loss'][0]
    loss_fns.append(loss)
    
    if loss < t_loss:
        converged = True
```

```python
plt.figure(1)
plt.plot(range(1, len(loss_fns) + 1), loss_fns)
plt.title(f'SSE vs. Number of Epochs (Converged at Epoch {len(loss_fns)})')
plt.xlabel('Epochs')
plt.ylabel('SSE')

plt.figure(2)
plt.plot(range(1, len(lr) + 1), lr)
plt.title('Learning Rate vs Number of Epochs')
plt.xlabel('Epochs')
plt.ylabel('Learning Rate')
```

```python

converged = False
t_loss = 0.001
loss_fns = []
lr = []
mv = 0.9

model = keras.Sequential()
model.add(Dense(8, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
learning_rate = 0.1
sgd2 = keras.optimizers.SGD(learning_rate=learning_rate, momentum=mv)
model.compile(optimizer=sgd2, loss='mean_squared_error')

class LearningRateRecorder(Callback):
    def on_epoch_end(self, epoch, logs=None):
        current_learning_rate = float(self.model.optimizer.learning_rate)
        lr.append(current_learning_rate)
        

while not converged:
    history = model.fit(data, target, epochs=10, verbose=0, callbacks=[LearningRateRecorder()])
    loss = history.history['loss'][0]
    loss_fns.append(loss)
    
    if loss < t_loss:
        converged = True
```

```python
plt.figure(3)
plt.plot(range(1, len(loss_fns) + 1), loss_fns)
plt.title(f'SSE vs. Number of Epochs (Converged at Epoch {len(loss_fns)})')
plt.xlabel('Epochs')
plt.ylabel('SSE')

plt.figure(4)
plt.plot(range(1, len(lr) + 1), lr)
plt.title('Learning Rate vs  Epochs')
plt.xlabel('Epochs')
plt.ylabel('Learning Rate')
```

```python
learning_rate_adaptive = 0.1
lr_adaptive = []
converged_adaptive = False
loss_fns_adaptive = []

model = keras.Sequential()
model.add(Dense(8, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
sgd3 = keras.optimizers.SGD(learning_rate=learning_rate_adaptive)
model.compile(optimizer=sgd3, loss='mean_squared_error')

class LearningRateRecorder(Callback):
    def on_epoch_end(self, epoch, logs=None):
        current_learning_rate = float(self.model.optimizer.learning_rate)
        lr_adaptive.append(current_learning_rate)
        
prev_loss, loss = float('inf'), float('inf')

while not converged_adaptive:
    history = model.fit(data, target, epochs=5, verbose=0, callbacks=[LearningRateRecorder()])
    prev_loss = loss
    loss = history.history['loss'][0]
    loss_fns_adaptive.append(loss)
    
    if loss > 1.04 * prev_loss:
        learning_rate_adaptive *= 0.7
    elif loss < prev_loss:
        learning_rate_adaptive *= 1.05

    optimizer = keras.optimizers.SGD(learning_rate=learning_rate_adaptive)
    model.compile(optimizer=optimizer, loss='mean_squared_error')

    if loss < t_loss:
        converged_adaptive = True
        break
```

```python
plt.figure(5)
plt.plot(range(1, len(loss_fns_adaptive) + 1), loss_fns_adaptive)
plt.title(f'SSE vs. Epoch (Converged at Epoch {len(loss_fns)})')
plt.xlabel('Epoch')
plt.ylabel('SSE')

plt.figure(6)
plt.plot(range(1, len(lr_adaptive) + 1), lr_adaptive)
plt.title('Learning Rate vs Epochs')
plt.xlabel('Epoch')
plt.ylabel('Learning Rate')
```
